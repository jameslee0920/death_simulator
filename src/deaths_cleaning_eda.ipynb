Looks like Population may need to be an integer, but let's see what kinds of values are in each column:

```python
for c in deaths[['Notes','State','Ten-Year Age Groups','Ten-Year Age Groups Code',
                 'Gender','Gender Code','Race','Race Code','Population','Crude Rate']].columns:
    print("---- %s ---" % c)
    print(deaths[c].value_counts())
```

"Unreliable" and "Not Applicable" are in crude rate, "Not Applicable" is in Population, and "Not Stated" is within Age Groups.  According to the documentation at [CDC Wonder](https://wonder.cdc.gov/controller/datarequest/D76;jsessionid=6989CE698643E7A64442C371308570C4), "Not Applicable" and "Unreliable" happens when there are 20 or fewer cases. This being the case, I think we're safe to remove without ruining the results. "Not Stated" for Age Groups means they are included in total counts and rates, but are not distributed among age groups.  This, too, can be removed.  

Finally, since several of the columns are duplicate "code" versions, we can do away with them.

```python
deaths = deaths[deaths['Population'] != "Not Applicable"]
deaths = deaths[deaths['Crude Rate'] != "Not Applicable"]
deaths = deaths[deaths['Crude Rate'] != "Unreliable"]
deaths = deaths[deaths['Ten-Year Age Groups'] != "Not Stated"]
```

```python
deaths.dtypes
```

```python
deaths = deaths[['Year','State','Ten-Year Age Groups','Gender','Race','Deaths','Population','Crude Rate']]
```

With Population and Crude Rate scrubbed of strings, we can convert to int and float.

```python
deaths['Population'] = deaths['Population'].astype(int)
deaths['Crude Rate'] = deaths['Crude Rate'].astype(float)
```

One more check on the value counts of the object columns:

```python
for c in deaths[['State','Ten-Year Age Groups','Gender','Race']].columns:
    print("---- %s ---" % c)
    print(deaths[c].value_counts())
```
